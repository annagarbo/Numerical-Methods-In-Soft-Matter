```python
import numpy as np

# Funzione g(x) corretta
def g(x, a, b):
    return a + b * x**2


# Funzione da integrare
def f(x):
    return np.cos(x)


a_opt=1
b_opt=(24/(np.pi)**3)*(1-np.pi/2)

# Numero di iterazioni
num_iterations = 0
desired_accuracy = 0.01  # 1% accuracy
estimated_integral = 0

while True:
    num_iterations += 1
    samples_y = np.random.uniform(0, 1, 10000)
    samples=np.sqrt((1-samples_y)*3/b_opt)
    weights = f(samples_y) / g(samples_y,a_opt,b_opt)
    estimated_integral = np.mean(weights)
    estimated_variance = np.var(weights) / num_iterations
    
    if np.sqrt(estimated_variance) < desired_accuracy * np.abs(estimated_integral):
        break

# Stampa il risultato
print("Stima dell'integrale:", estimated_integral)
print("Numero di iterazioni:", num_iterations)
print("Estimated Variance:",estimated_variance)
```
