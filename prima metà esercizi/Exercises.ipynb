```python
data_list = []  # Creare una lista per memorizzare i dati

with open("hit_miss.csv", newline='') as csvfile:
    for line in csvfile:
        data = line.strip().split()
        if len(data) == 3:
            x, y, r = map(float, data)
            data_list.append([x, y, r])  # Aggiungi i dati a data_list

# Creare un DataFrame di Pandas
df = pd.DataFrame(data_list, columns=["x", "y", "r"])
df
# Ora df contiene i dati in un DataFrame con le colonne x, y e r

            
```

```python
plt.scatter(df['x'],df['y'],s=5,alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
area_ret=len(df)/10000000*(5.-2.)*(3.-1.)
print('Area estimated with montecarlo simulation:',area_ret)
print('Area estimated analitically:',(5.-2.)*(3.-1.))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leggi il file CSV in un DataFrame
df = pd.read_csv("power_law_points_3.csv")

# Estrai la colonna dei dati
data = df

# Crea un istogramma
plt.hist(data, bins=30, density=True ,alpha=0.7, color='blue', label="Generated data")

# Crea una curva teorica di power law
x_values = np.linspace(0.001, 1, 100)  # Evita divisione per zero
n = 3  # Valore di n per la power law
y_values = 4*x_values**(n )

# Sovrapponi la curva teorica all'istogramma
plt.plot(x_values, y_values, 'r-', label=f"Power Law (n={n})")

plt.title("Comparison between generate data and Power Law")
plt.xlabel("X values")
plt.ylabel("Probability density")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#p=1
rej=pd.read_csv('reject1.csv')
def target_pdf(x):
    return 2/np.sqrt(np.pi) * np.exp(-x**2)




# Create a histogram to compare the generated samples to the target PDF
plt.figure(figsize=(8, 6))
plt.hist(rej, bins=30, density=True, alpha=1, label='Generated Samples',color='blue')
x = np.linspace(0, 3, 1000)
plt.plot(x, target_pdf(x), 'r', label='Target PDF: f(x)')

plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Rejection Method for Generating Samples')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt



def func(x, p, A):
    result = np.empty_like(x)
    result[x <= p] = A
    result[x > p] = (A / p) * x[x > p] * np.exp(p**2 - x[x > p]**2)
    return result



    
    
p=0.7
A = 2*p/(1+2*p**2)
x = np.arange(0, 3.1459/2, 0.1)
plt.plot(x, np.sqrt(2/np.pi) * np.exp(-x**2))
plt.plot(x, 1*f(x,p,A))
plt.show()
```
