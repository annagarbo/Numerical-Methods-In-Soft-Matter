```python
import numpy as np


P2 = np.array([[0, 0, 1], [0, 1, 0], [1/4, 0, 3/4]])

for n in range(1, 6):
    Pn = np.linalg.matrix_power(P2, n)
    print(f'P^{n}:\n{Pn}\n')


limit_inf = np.linalg.matrix_power(P2, 100000)  
print(f'Limite mentre n -> ∞:\n{limit_inf}\n')

```

```python
import numpy as np

def is_regular(matrix):
    num_states = matrix.shape[0]
    identity_matrix = np.identity(num_states)
    transition_matrix = matrix.copy()

    for _ in range(100):  # Limit the number of iterations for safety
        
        transition_matrix = np.dot(transition_matrix, matrix)
        
        if np.allclose(transition_matrix, identity_matrix):
            return True
        else:
            return False

# Definisci la matrice di transizione P
P = np.array([[1/2, 1/4, 1/4],
              [1/2, 0,   1/2],
              [1/4, 1/4, 1/2]])

for n in range(1, 6):
    Pn = np.linalg.matrix_power(P, n)
    print(f'P^{n}:\n{Pn}\n')

# Calcola il limite mentre n -> ∞
limit_inf = np.linalg.matrix_power(P, 100000)  # Utilizzo di un valore grande come approssimazione di ∞
print(f'Limite mentre n -> ∞:\n{limit_inf}\n')


```

```python
import numpy as np
p=0.2

P = np.array([[p, 1-p, 0, 0],
              [0, 0  , p, 1-p],
              [p, 1-p, 0, 0],
              [0, 0  , p, 1-p]
             ])


eigenvalues, eigenvectors = np.linalg.eig(P.T)  
stationary_distribution = np.real(eigenvectors[:, 0] / np.sum(eigenvectors[:, 0]))

print("Stationary Distribution:")
print(stationary_distribution)




```

```python
import numpy as np
from sympy import symbols, solve

# Definisci il simbolo p
p = symbols('p')

# Definisci la matrice di transizione P in termini del simbolo p
P = np.array([[p, 1-p, 0, 0],
              [0, 0, p, 1-p],
              [p, 1-p, 0, 0],
              [0, 0, p, 1-p]
             ])

# Risolvi il sistema di equazioni lineari per trovare l'autovettore in funzione di p
eigenvalues, eigenvectors = np.linalg.eig(P.T)
stationary_distribution_symbolic = eigenvectors[:, 0] / np.sum(eigenvectors[:, 0])

# Risolvi il sistema di equazioni lineari per ottenere l'autovettore in funzione di p
solution = solve(stationary_distribution_symbolic - stationary_distribution_symbolic.sum(), p)

# Stampa la soluzione
print("Autovettore in funzione di p:")
print(solution)
```
