```python
def Read_coor(fn,N,skip):
    s = np.loadtxt(fn, delimiter=" ",skiprows=1+(N+2)*skip,max_rows=N+1,dtype=str)
    xy = (s.T).astype(float)
    return xy

csmall="#5577EE"

def Show_conf(ax, xy, cp=None, title="",s=40, bond=False):
    #v=np.array(((0,0),(Lx,0),(Lx,Ly),(0,Ly),(0,0)))
    #ax.plot(v.T[0],v.T[1],c="k")
    if cp==None: c=csmall
    else:        c=cp
    x,y=xy[0],xy[1]
    b=sqrt(np.power(x[1:]-x[:-1],2)+np.power(y[1:]-y[:-1],2))
    bb=np.where(b<6)
    ab=np.zeros(len(b))
    ab[bb]=0.2
    Y=50
    #print(x[:Y])
    #print(y[:Y])
    PRI=False
    if PRI:
        print("bonds ",b[:Y])
        print("ind. short ones: ",bb)
        print("alpha : ",ab)
    circle0 = plt.Circle((x[0],y[0]), 2*(R0+R), color="#EEBB33",alpha=1)
    ax.add_patch(circle0)
    ax.scatter(x[1:-1],y[1:-1],c=c[1:-1],marker="o",linewidths=0,s=s,alpha=0.2)
    # BONDS
    if bond:
        for m in range(Mp):
            i=1+m*Lp
            ii=i+Lp
            ax.plot(x[i:ii],y[i:ii],c=c[i],alpha=0.2)
    #    ax.scatter(np.mod(x[-1],Lx),Ly/2,c="#FF1100",marker="x",linewidths=3,s=140,alpha=0.65)
    ax.scatter(x[-1],y[-1],c="#FF1100",marker="x",linewidths=3,s=140,alpha=0.65)
    ax.axis('off')
    ax.set_title(title)
```
