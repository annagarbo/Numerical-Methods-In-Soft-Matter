```python
import numpy as np
import matplotlib.pyplot as plt

# Carica i dati dai file di testo
def load_data(filename):
    data = np.loadtxt(filename)
    return data[:, 0], data[:, 1]

# Carica i dati
T, E = load_data('energy.txt')
T, M = load_data('magnetization.txt')
T, C = load_data('specific_heat.txt')
T, X = load_data('susceptibility.txt')

# Creazione della figura e degli assi
fig, axs = plt.subplots(2, 2, figsize=(18, 10))

# Grafico dell'energia
axs[0, 0].scatter(T, E, s=50, marker='o', color='IndianRed')
axs[0, 0].set_xlabel('Temperature (T)', fontsize=20)
axs[0, 0].set_ylabel('Energy', fontsize=20)
axs[0, 0].set_title('Energy vs Temperature', fontsize=20)
axs[0, 0].axis('tight')

# Grafico della magnetizzazione
axs[0, 1].scatter(T, M, s=50, marker='o', color='RoyalBlue')
axs[0, 1].set_xlabel('Temperature (T)', fontsize=20)
axs[0, 1].set_ylabel('Magnetization', fontsize=20)
axs[0, 1].set_title('Magnetization vs Temperature', fontsize=20)
axs[0, 1].axis('tight')

# Grafico della capacit termica
axs[1, 0].scatter(T, C, s=50, marker='o', color='IndianRed')
axs[1, 0].set_xlabel('Temperature (T)', fontsize=20)
axs[1, 0].set_ylabel('Specific Heat', fontsize=20)
axs[1, 0].set_title('Specific Heat vs Temperature', fontsize=20)
axs[1, 0].axis('tight')

# Grafico della suscettivit
axs[1, 1].scatter(T, X, s=50, marker='o', color='RoyalBlue')
axs[1, 1].set_xlabel('Temperature (T)', fontsize=20)
axs[1, 1].set_ylabel('Susceptibility', fontsize=20)
axs[1, 1].set_title('Susceptibility vs Temperature', fontsize=20)
axs[1, 1].axis('tight')

# Mostra i grafici
plt.tight_layout()
plt.show()
```
