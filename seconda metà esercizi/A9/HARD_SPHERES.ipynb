```python
import matplotlib.pyplot as plt
import numpy as np

# Dati estratti
dmax_values = np.array([0.01, 0.2575, 0.505, 0.7525])
densities = np.array([0.01, 0.03, 0.05, 0.1])

# Dati per Random Init
acceptance_ratios_random = np.array([
    [0.00, 0.00, 0.00, 0.00],  # Density: 0.01
    [0.00, 0.00, 0.00, 0.00],  # Density: 0.03
    [0.00, 0.00, 0.00, 0.00],  # Density: 0.05
    [0.00, 0.00, 0.00, 0.00]   # Density: 0.10
])
energies_random = np.array([
    [1e30, 1e30, 1e30, 1e30],  # Density: 0.01
    [1e30, 1e30, 1e30, 1e30],  # Density: 0.03
    [1e30, 1e30, 1e30, 1e30],  # Density: 0.05
    [1e30, 1e30, 1e30, 1e30]   # Density: 0.10
])

# Dati per Lattice Init
acceptance_ratios_lattice = np.array([
    [0.10, 0.10, 0.10, 0.10],  # Density: 0.01
    [0.09, 0.09, 0.09, 0.09],  # Density: 0.03
    [0.08, 0.08, 0.08, 0.08],  # Density: 0.05
    [0.07, 0.07, 0.07, 0.07]   # Density: 0.10
])
energies_lattice = np.array([
    [1.58e15, 1.68e15, 1.22e15, 1.72e15],  # Density: 0.01
    [1.22e16, 1.23e16, 1.21e16, 1.19e16],  # Density: 0.03
    [2.32e16, 2.21e16, 2.45e16, 2.34e16],  # Density: 0.05
    [3.34e16, 3.33e16, 3.26e16, 3.49e16]   # Density: 0.10
])

# Funzione per creare i grafici
def plot_results():
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))

    # Grafico 1: Energia vs dmax per Random Init
    axs[0, 0].plot(dmax_values, energies_lattice[0], label='Density 0.01', marker='o')
    axs[0, 0].plot(dmax_values, energies_lattice[1], label='Density 0.03', marker='o')
    axs[0, 0].plot(dmax_values, energies_lattice[2], label='Density 0.05', marker='o')
    axs[0, 0].plot(dmax_values, energies_lattice[3], label='Density 0.10', marker='o')
    axs[0, 0].set_xlabel('dmax')
    axs[0, 0].set_ylabel('Average Energy')
    axs[0, 0].set_title('Lattice Init: Average Energy vs dmax')
    axs[0, 0].legend()

    # Grafico 2: Rapporto di Accettazione vs dmax per Random Init
    axs[0, 1].plot(dmax_values, acceptance_ratios_lattice[0], label='Density 0.01', marker='o')
    axs[0, 1].plot(dmax_values, acceptance_ratios_lattice[1], label='Density 0.03', marker='o')
    axs[0, 1].plot(dmax_values, acceptance_ratios_lattice[2], label='Density 0.05', marker='o')
    axs[0, 1].plot(dmax_values, acceptance_ratios_lattice[3], label='Density 0.10', marker='o')
    axs[0, 1].set_xlabel('dmax')
    axs[0, 1].set_ylabel('Acceptance Ratio')
    axs[0, 1].set_title('Lattice Init: Acceptance Ratio vs dmax')
    axs[0, 1].legend()

    # Grafico 3: Energia vs densit per Lattice Init
    axs[1, 0].plot(densities, energies_lattice[:, 0], label='dmax 0.01', marker='o')
    axs[1, 0].plot(densities, energies_lattice[:, 1], label='dmax 0.2575', marker='o')
    axs[1, 0].plot(densities, energies_lattice[:, 2], label='dmax 0.505', marker='o')
    axs[1, 0].plot(densities, energies_lattice[:, 3], label='dmax 0.7525', marker='o')
    axs[1, 0].set_xlabel('Density')
    axs[1, 0].set_ylabel('Average Energy')
    axs[1, 0].set_title('Lattice Init: Average Energy vs Density')
    axs[1, 0].legend()

    # Grafico 4: Rapporto di Accettazione vs densit per Lattice Init
    axs[1, 1].plot(densities, acceptance_ratios_lattice[:, 0], label='dmax 0.01', marker='o')
    axs[1, 1].plot(densities, acceptance_ratios_lattice[:, 1], label='dmax 0.2575', marker='o')
    axs[1, 1].plot(densities, acceptance_ratios_lattice[:, 2], label='dmax 0.505', marker='o')
    axs[1, 1].plot(densities, acceptance_ratios_lattice[:, 3], label='dmax 0.7525', marker='o')
    axs[1, 1].set_xlabel('Density')
    axs[1, 1].set_ylabel('Acceptance Ratio')
    axs[1, 1].set_title('Lattice Init: Acceptance Ratio vs Density')
    axs[1, 1].legend()

    plt.tight_layout()
    plt.show()

# Creazione dei grafici
plot_results()
```
