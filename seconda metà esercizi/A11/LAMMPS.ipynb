```python
import pandas as pd

# Set density and a constant for the virial expansion
density = 200/1000
b = -0.772

# Read simulation data into a DataFrame
filename = 'berendsen_200.txt'
data = pd.read_csv(filename, delim_whitespace=True, header=0)

# Calculate average kinetic energy, temperature, and pressure
mean_kinetic_energy = data['KinEng'].mean()
mean_temperature = data['Temp'].mean()
mean_pressure = data['Press'].mean()

# Calculate theoretical kinetic energy based on equipartition theorem
theoretical_kinetic_energy = 1.5 * mean_temperature

# Calculate theoretical pressure using virial expansion
theoretical_pressure = density * mean_temperature + mean_temperature * density**2 * b


# Compare calculated and theoretical values
if abs((mean_kinetic_energy - theoretical_kinetic_energy) / theoretical_kinetic_energy) < 0.02:
    print("The average kinetic energy is consistent with the equipartition theorem.")
else:
    print("The average kinetic energy is not consistent with the equipartition theorem.")
if abs((mean_pressure - theoretical_pressure) ) < 0.02:
    print("The average pressure is consistent with the virial expansion.")
else:
    print("The average pressure is not consistent with the virial expansion.")

print('Mean Kinetic Energy:',round(mean_kinetic_energy,2))
print('Theoretical Kinetic Energy:',round(theoretical_kinetic_energy,2))
print('Mean pressure:',round(mean_pressure,2))
print('Theoretical pressure:',round(theoretical_pressure,2))
```
