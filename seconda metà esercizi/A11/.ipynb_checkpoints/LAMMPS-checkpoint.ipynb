```python
import numpy as np
import matplotlib.pyplot as plt

# Leggere il file rdf.dat
filename = 'rdf.dat'
data = []

with open(filename, 'r') as file:
    for line in file:
        # Ignorare le righe di commento e intestazione
        if line.startswith('#') or line.startswith('Time'):
            continue
        # Dividere la linea in colonne e convertire in float
        data.append([float(x) for x in line.split()])

# Convertire i dati in un array numpy per una facile manipolazione
data = np.array(data)
print(data)
# La prima colonna è r, la seconda è g(r)
r = data[:, 1]  # la prima colonna di dati inizia da colonna 1
gr = data[:, 2]  # la seconda colonna di dati inizia da colonna 2

# Plot della funzione di distribuzione radiale g(r)
plt.figure(figsize=(8, 6))
plt.plot(r, gr, label='g(r)')
plt.xlabel('r (distanza)')
plt.ylabel('g(r)')
plt.title('Funzione di Distribuzione Radiale (g(r))')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Leggere il file energy.dat
filename = 'energy.dat'
data = []

with open(filename, 'r') as file:
    for line in file:
        # Ignorare le righe di commento e intestazione
        if line.startswith('Step'):
            continue
        # Dividere la linea in colonne e convertire in float
        data.append([float(x) for x in line.split()])

# Convertire i dati in un array numpy per una facile manipolazione
data = np.array(data)
print(data)

# La prima colonna è il passo, la seconda è l'energia potenziale, la terza è l'energia totale
steps = data[:, 0]
pe = data[:, 1]  # Energia potenziale
etotal = data[:, 2]  # Energia totale

# Plot delle energie in funzione dei passi della simulazione
plt.figure(figsize=(10, 6))
plt.plot(steps, pe, label='Energia Potenziale (PE)')
plt.plot(steps, etotal, label='Energia Totale (ETOTAL)', linestyle='--')
plt.xlabel('Passi della Simulazione')
plt.ylabel('Energia')
plt.title('Energia in Funzione dei Passi della Simulazione')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Lista dei valori di cutoff sigma
cutoff_values = [
    1.12246205, 1.32246205, 1.52246205, 1.72246205, 1.92246205,
    2.12246205, 2.32246205, 2.52246205, 2.72246205, 2.92246205,
    3.12246205, 3.32246205, 3.52246205, 3.72246205, 3.92246205
]

# Creare un colore differente per ogni grafico
colors = plt.cm.viridis(np.linspace(0, 1, len(cutoff_values)))

plt.figure(figsize=(12, 8))

for i, sigma in enumerate(cutoff_values):
    filename = f'rdf{i+1}_output.dat'
    data = []

    try:
        with open(filename, 'r') as file:
            for line in file:
                # Ignorare le righe di commento e intestazione
                if line.startswith('#') or line.startswith('Time'):
                    continue
                # Dividere la linea in colonne
                split_line = line.split()
                # Considerare solo le righe con il numero corretto di colonne
                if len(split_line) == 4:  # Assumendo che ci siano 4 colonne di dati
                    data.append([float(x) for x in split_line])

        # Convertire i dati in un array numpy per una facile manipolazione
        data = np.array(data)

        # Assicurarsi che i dati siano stati caricati correttamente
        if data.size == 0:
            print(f"Avviso: Il file {filename} è vuoto o non valido.")
            continue

        # La prima colonna è r, la seconda è g(r)
        r = data[:, 1]  # seconda colonna del file
        gr = data[:, 2]  # terza colonna del file

        # Plot della funzione di distribuzione radiale g(r)
        plt.scatter(r, gr, color=colors[i], label=f'sigma = {sigma:.2f}',alpha=0.5)
    
    except FileNotFoundError:
        print(f"Errore: Il file {filename} non è stato trovato.")

plt.xlabel('r (distanza)')
plt.ylabel('g(r)')
plt.title('Funzione di Distribuzione Radiale (g(r)) per diversi valori di cutoff')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

d=200/1000
b=-0.772
# Leggere il file e creare un DataFrame
filename = 'berendsen_200.txt'
df = pd.read_csv(filename, delim_whitespace=True, header=0)

# Calcolare l'energia cinetica media
mean_kinetic_energy = df['KinEng'].mean()

# Leggere la temperatura media
mean_temperature = df['Temp'].mean()

# Calcolare l'energia cinetica teorica basata sull'equipartizione
# In unità LJ: k_B = 1, quindi E_K = 3/2 * T
theoretical_kinetic_energy = 1.5 * mean_temperature

#pressione media
mean_pressure=df['Press'].mean()
#pressione teorica
th_press=d*mean_temperature+mean_temperature*d*d*b

print(f"Energia cinetica media calcolata: {mean_kinetic_energy:.5f}")
print(f"Energia cinetica teorica basata sull'equipartizione: {theoretical_kinetic_energy:.5f}")

# Verificare se l'energia cinetica media è consistente con la teoria
if abs((mean_kinetic_energy - theoretical_kinetic_energy)/theoretical_kinetic_energy) < 0.02:
    print("L'energia cinetica media è consistente con il teorema dell'equipartizione.")
else:
    print("L'energia cinetica media non è consistente con il teorema dell'equipartizione.")

# Verificare se l'energia cinetica media è consistente con la teoria
if abs((mean_pressure - th_press)/th_press) < 0.02:
    print("La pressione media è consistente con la virial expansion.")
else:
    print("La pressione media è consistente con la virial expansion.")
```
